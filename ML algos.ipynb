```python
import pandas as pd 
import numpy as np
```

**Dataset Preprocessing**

```python
df=pd.read_csv("IOT_Assignment_2_data_regression_sensor_range.csv")
```

```python
df
```

```python
min_humidity=0
max_humidity=100
df['Humidity(%)']=(df['Humidity(%)']-min_humidity)/(max_humidity-min_humidity)
```

```python
min_temp=-40
max_temp=100
df['Temperature(°C)']=(df['Temperature(°C)']-min_temp)/(max_temp-min_temp)
```

```python
'''
waterflow=MinMaxScaler()
df['WaterFlow(%)']=waterflow.fit_transform(df['WaterFlow(%)'].values.reshape(-1,1))
'''
```

Dataset Split

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=100,test_size=0.2)
```

**Model**

1 Neural Network

Input Layer->Hidden1->Hidden2->Output Later

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# define the keras model
model = Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# fit the keras model on the dataset
model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

2 Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rt_reg=RandomForestRegressor(max_depth=4,n_estimators=10,random_state=0)
rt_reg.fit(x_train,y_train)
```

3 Linear Regression

```python
from sklearn import linear_model,datasets
model_LR = linear_model.LinearRegression(normalize=True)
model_LR.fit(x_train,y_train)
```

4 Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg= DecisionTreeRegressor(max_depth=10,random_state=0)
dt_reg.fit(x_train,y_train)
```

5 SV Regression

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
```

**Evaluation**

Mean Square Error

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
y_nn=model.predict(x_test)
y_rt=rt_reg.predict(x_test)
y_lr=model_LR.predict(x_test)
y_dt=dt_reg.predict(x_test)
y_svr=svr.predict(x_test)
```

```python
print("RMSE of Neural Network",mse(y_nn,y_test,squared=False))
print("RMSE of Random Forest",mse(y_rt,y_test,squared=False))
print("RMSE of Linear Regression",mse(y_lr,y_test,squared=False))
print("RMSE of Decision Tree",mse(y_dt,y_test,squared=False))
print("RMSE of Support Vector",mse(y_svr,y_test,squared=False))
```

From these observation we have seen that Decision Tree perform optimal in Testing Set. 

So we are taking Random Forest as our model
